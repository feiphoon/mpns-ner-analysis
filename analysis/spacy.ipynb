```python
import os
import spacy
from spacy import displacy

# nlp = spacy.load("en_core_web_sm",  disable=["tagger", "parser", "attribute_ruler", "lemmatizer"])
```

```python
daisy_data_filepath = "../data/daisy"

daisy_data_filenames = os.listdir(daisy_data_filepath)

daisy_data_filenames
```

```python
def run_ner_model(model: str, data_filepath: str, data_filenames: str) -> None:
    nlp = spacy.load(model)

    for _ in data_filenames:
        with open(os.path.join(data_filepath, _), "r") as f:
            text = f.read()
            doc = nlp(text)
            displacy.render(doc, style="ent")
            print("\n")
```

```python
run_ner_model("en_core_web_sm", daisy_data_filepath, daisy_data_filenames)
```

```python
run_ner_model("en_core_web_lg", daisy_data_filepath, daisy_data_filenames)
```

```python
run_ner_model("en_core_web_trf", daisy_data_filepath, daisy_data_filenames)
```
